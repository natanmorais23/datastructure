```python
import networkx as nx

# Ler o grafo
G = nx.read_gexf('/content/network.gexf')

# Remover self-loops, se existirem
G.remove_edges_from(nx.selfloop_edges(G))

# Calcular o k-core
k_core_dict = nx.core_number(G)

# Adicionar o atributo "k-core"
nx.set_node_attributes(G, k_core_dict, "k-core")

# Definir os grupos de interesse
core_value = 17  # Valor do núcleo a ser destacado
shell_value = 13  # Valor da camada a ser destacada

# Criar um atributo categórico para os grupos
group_dict = {}
for node in G.nodes():
    if k_core_dict[node] == core_value:  # Grupo do núcleo
        group_dict[node] = "17-core"
    elif k_core_dict[node] == shell_value:  # Grupo da camada
        group_dict[node] = "13-shell"
    else:  # Restante
        group_dict[node] = "others"

# Adicionar o atributo "group"
nx.set_node_attributes(G, group_dict, "group")

# Exportar o grafo para o Gephi
output_path = "grafo_com_grupos.gexf"
nx.write_gexf(G, output_path)

print(f"Grafo com grupos categóricos exportado para {output_path}!")
```
