```python
import pandas as pd

# Substitua pelo caminho correto até o arquivo CSV no seu Drive
file_path = '/content/drive/My Drive/U1T3 - AED2/DADOS_ABERTOS_MEDICAMENTOS.csv'

# Carregar o arquivo CSV
df = pd.read_csv(file_path, delimiter=';', encoding='ISO-8859-1')

# Exibir as primeiras linhas do dataframe
print(df.head())
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations

# Criar um grafo
G = nx.Graph()

# Adicionar nós: cada medicamento é um nó
for idx, row in df.iterrows():
    G.add_node(row['NOME_PRODUTO'],
               empresa=row['EMPRESA_DETENTORA_REGISTRO'],
               classe=row['CLASSE_TERAPEUTICA'],
               principio_ativo=row['PRINCIPIO_ATIVO'])

# Criar arestas entre medicamentos que compartilham pelo menos um princípio ativo
principio_ativos_medicamentos = df.groupby('PRINCIPIO_ATIVO')['NOME_PRODUTO'].apply(list)

for medicamentos in principio_ativos_medicamentos:
    for med1, med2 in combinations(medicamentos, 2):
        G.add_edge(med1, med2)

# Calcular a assortatividade por empresa
assortatividade_empresa = nx.attribute_assortativity_coefficient(G, 'empresa')
print(f"Assortatividade por Empresa: {assortatividade_empresa}")

# Calcular a assortatividade por classe terapêutica
assortatividade_classe = nx.attribute_assortativity_coefficient(G, 'classe')
print(f"Assortatividade por Classe Terapêutica: {assortatividade_classe}")

# Conclusão:
if assortatividade_empresa > 0:
    print(f"Medicamentos da mesma empresa tendem a compartilhar mais princípios ativos (Assortatividade = {assortatividade_empresa}).")
else:
    print(f"Medicamentos da mesma empresa não tendem a compartilhar mais princípios ativos (Assortatividade = {assortatividade_empresa}).")

if assortatividade_classe > 0:
    print(f"Medicamentos da mesma classe terapêutica tendem a compartilhar mais princípios ativos (Assortatividade = {assortatividade_classe}).")
else:
    print(f"Medicamentos da mesma classe terapêutica não tendem a compartilhar mais princípios ativos (Assortatividade = {assortatividade_classe}).")
```
