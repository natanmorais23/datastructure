```python
df = pd.read_csv('/content/drive/MyDrive/AED/DADOS_ABERTOS_MEDICAMENTOS.csv', encoding='latin1', delimiter=';')
```

```python
import pandas as pd
import networkx as nx

# Supondo que o dataset está carregado em um DataFrame chamado `df`
# df = pd.read_csv('seu_dataset.csv') # Carregue o dataset se necessário

# Filtrar os dados relevantes
df = df[df['PRINCIPIO_ATIVO'].notna()]  # Remover medicamentos sem princípio ativo

# Criar a rede
G = nx.Graph()

# Adicionar nós com o nome do produto como rótulo e categoria regulatória como atributo
for idx, row in df.iterrows():
    G.add_node(row['NOME_PRODUTO'], categoria_regulatoria=row['CATEGORIA_REGULATORIA'])

# Criar arestas para medicamentos que compartilham princípios ativos
# Primeiro, agrupar medicamentos pelo princípio ativo
principio_ativo_groups = df.groupby('PRINCIPIO_ATIVO')

for _, group in principio_ativo_groups:
    medicamentos = group['NOME_PRODUTO'].tolist()
    for i in range(len(medicamentos)):
        for j in range(i + 1, len(medicamentos)):
            G.add_edge(medicamentos[i], medicamentos[j])

# Calcular a assortatividade com base na categoria regulatória
assortatividade_categoria = nx.attribute_assortativity_coefficient(G, 'categoria_regulatoria')
print(f"Assortatividade pela categoria regulatória: {assortatividade_categoria}")

# Responder a pergunta
if assortatividade_categoria > 0:
    print("Medicamentos da mesma categoria regulatória tendem a compartilhar princípios ativos.")
else:
    print("Medicamentos de diferentes categorias regulatórias tendem a compartilhar princípios ativos.")
```
