```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import nxviz as nv
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/AED/scopus.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
G = nx.Graph()
```

```python
for authors in df['Authors']:
    # Separando os autores e removendo espaços extras
    author_list = [author.strip() for author in authors.split(';')]

    # Adicionando nós e arestas ao grafo
    for i in range(len(author_list)):
        for j in range(i + 1, len(author_list)):
            G.add_edge(author_list[i], author_list[j])

```

```python
G.nodes(data=True)
```

```python
densidade_rede = nx.density(G)
print("Densidade da rede:", densidade_rede)
```

```python
X = 3  # Defina o valor de X aqui

# Obtenha nós com grau >= X
nodos_filtrados = [n for n, d in G.degree() if d >= X]

# Crie o sub-grafo com base nos nós filtrados
sub_grafo = G.subgraph(nodos_filtrados)

# Calcule a densidade do sub-grafo
densidade_sub_grafo = nx.density(sub_grafo)
print("Densidade do sub-grafo:", densidade_sub_grafo)
```

```python
import matplotlib.pyplot as plt

# Obtenha a lista dos graus de cada nó
graus = [d for n, d in G.degree()]

# Plote o histograma dos graus
plt.hist(graus, bins=range(1, max(graus) + 2), edgecolor="black", align="left")
plt.title("Histograma de Graus")
plt.xlabel("Grau (Número de coautores)")
plt.ylabel("Frequência")
plt.show()
```
