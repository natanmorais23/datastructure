# Capítulo 02 - Questão 01

```python
probabilidades = [
    (0.5, 0.5),
    (0.6, 0.7),
    (0.4, 0.8),
    (0.1, 0.2),
    (0.3, 0.4)
]

def calcular_probabilidades(p1_H, p2_H):
    p1_T = 1 - p1_H
    p2_T = 1 - p2_H

    HH = p1_H * p2_H
    HT = p1_H * p2_T
    TH = p1_T * p2_H
    TT = p1_T * p2_T

    return HH, HT, TH, TT

resultados = []
for p1_H, p2_H in probabilidades:
    resultados.append(calcular_probabilidades(p1_H, p2_H))

resultados
```

# Capítulo 02 - Questão 03

```python
def caesar_cipher_shift(text, shift=7):
    result = []
    for char in text:
        if char.isalpha():
            start = ord('A') if char.isupper() else ord('a')
            result.append(chr(start + (ord(char) - start + shift) % 26))
        else:
            result.append(char)
    return ''.join(result)

cipher_text = "OCZ XJMMZXO VINRZM"

decoded_text_step1 = caesar_cipher_shift(cipher_text, 7)
decoded_text_step2 = caesar_cipher_shift(decoded_text_step1, 7)
decoded_text_final = caesar_cipher_shift(decoded_text_step2, 7)

decoded_text_final
```

```python
import pandas as pd
import networkx as nx


edges = [
    ('A1', 'B1'),
    ('A1', 'B2'),
    ('A2', 'B2'),
    ('A2', 'B3'),
    ('A3', 'B1'),
    ('A3', 'B3'),
    ('A4', 'B4')
]

B = nx.Graph()
B.add_edges_from(edges)

nodes_type1 = {n for n in B.nodes if n.startswith('A')}  # Tipo 1 (por exemplo, A1, A2, ...)
nodes_type2 = {n for n in B.nodes if n.startswith('B')}  # Tipo 2 (por exemplo, B1, B2, ...)

max_neighbors_type1 = max(nodes_type1, key=lambda n: B.degree(n))
max_neighbors_type2 = max(nodes_type2, key=lambda n: B.degree(n))

print("Nodos do Tipo 1:", nodes_type1)
print("Nodos do Tipo 2:", nodes_type2)
print("Nó do Tipo 1 com mais vizinhos:", max_neighbors_type1, "com", B.degree(max_neighbors_type1), "vizinhos")
print("Nó do Tipo 2 com mais vizinhos:", max_neighbors_type2, "com", B.degree(max_neighbors_type2), "vizinhos")
```

# Capítulo 04 - Questão 02

```python
import pandas as pd
import networkx as nx

data = {
    'source': ['A1', 'A1', 'A2', 'B1', 'B2'],
    'target': ['B1', 'B2', 'B2', 'C1', 'C2'],
    'edge_type': [1, 1, 2, 'C', 'C']
}
df = pd.DataFrame(data)

layers = df['edge_type'].unique()
layer_graphs = {layer: nx.Graph() for layer in layers if isinstance(layer, int)}

for layer in layers:
    if isinstance(layer, int):
        sub_df = df[df['edge_type'] == layer]
        layer_graphs[layer].add_edges_from(zip(sub_df['source'], sub_df['target']))

is_star = any(layer_graph.number_of_edges() == layer_graph.number_of_nodes() - 1 for layer_graph in layer_graphs.values())
is_clique = all(layer_graph.number_of_edges() == layer_graph.number_of_nodes() * (layer_graph.number_of_nodes() - 1) // 2 for layer_graph in layer_graphs.values())
is_chain = all(layer_graph.number_of_edges() == layer_graph.number_of_nodes() - 1 for layer_graph in layer_graphs.values())

inter_layer_edges = df[df['edge_type'] == 'C']

if is_star:
    print("A rede contém um acoplamento tipo estrela.")
elif is_clique:
    print("A rede contém um acoplamento tipo clique.")
elif is_chain:
    print("A rede contém um acoplamento tipo cadeia.")
else:
    print("A rede não contém um acoplamento tipo estrela, clique ou cadeia.")

print("Número de arestas entre camadas:", inter_layer_edges.shape[0])
```

```python
import pandas as pd
import networkx as nx

data = [
    ['A1', 'A2', 'A3'],
    ['B1', 'B2'],
    ['C1', 'C2', 'C3', 'C4'],
]

unipartite_graph = nx.Graph()

for hyperedge in data:
    for i in range(len(hyperedge)):
        for j in range(i + 1, len(hyperedge)):
            unipartite_graph.add_edge(hyperedge[i], hyperedge[j])

bipartite_graph = nx.Graph()

for index, hyperedge in enumerate(data):
    hyperedge_node = f"Hyperedge_{index+1}"
    bipartite_graph.add_node(hyperedge_node, bipartite=1)  # Tipo 1: Hiperedge

    for node in hyperedge:
        bipartite_graph.add_node(node, bipartite=0)  # Tipo 0: Nó original
        bipartite_graph.add_edge(hyperedge_node, node)

print("Unipartite Graph Edges:")
print(unipartite_graph.edges())

print("\nBipartite Graph Edges:")
print(bipartite_graph.edges())
```

# Capítulo 04 - Questão 04

```python
import pandas as pd
import networkx as nx

data = {
    'source': ['A', 'A', 'B', 'C', 'C', 'D'],
    'target': ['B', 'C', 'C', 'D', 'E', 'E'],
    'first_snapshot': [1, 2, 1, 1, 3, 2],
    'last_snapshot': [3, 5, 3, 3, 4, 4]
}
df = pd.DataFrame(data)

window_size = 3
aggregated_edges = []

max_snapshot = df['last_snapshot'].max()
for start in range(1, max_snapshot + 1, window_size):
    end = start + window_size - 1
    current_edges = set()

    for _, row in df.iterrows():
        if row['first_snapshot'] <= end and row['last_snapshot'] >= start:
            current_edges.add((row['source'], row['target']))

    aggregated_edges.extend(current_edges)

aggregated_graph = nx.Graph()
aggregated_graph.add_edges_from(aggregated_edges)

print("Edges in the Aggregated Graph:")
print(aggregated_graph.edges())
```
